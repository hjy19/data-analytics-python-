### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users['is_male'] = np.where(users['gender'] == 'M', 1, 0)
male_ratio_per_occupation = users['is_male'].groupby(users['occupation']).mean()
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)
print(male_ratio_per_occupation)

```

### 31. For each occupation, calculate the minimum and maximum ages

```python
min_max_ages = users['age'].groupby(users['occupation']).agg(['min', 'max'])
print(min_max_ages)
```

### 32. For each combination of occupation and gender, calculate the mean age

```python
mean_age_by_occupation_and_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age_by_occupation_and_gender)
```

### 33.  For each occupation present the percentage of women and men

```python
gender_counts = users.groupby('occupation')['gender'].value_counts().reset_index(name='count')
occupation_counts = users.groupby('occupation')['gender'].count()
percentage_women_men = gender_counts.groupby('occupation').apply(lambda x: (x['count'] / occupation_counts[x['occupation'].iloc[0]]) * 100)
print(percentage_women_men)
```

### 49. Write a Python script to display the
- a. Current date and time
- b. Current year
- c. Month of year
- d. Week number of the year
- e. Weekday of the week
- f. Day of year
- g. Day of the month
- h. Day of week

```python
import datetime

current_datetime = datetime.datetime.now()

print("Current Date and Time:", current_datetime)

current_year = current_datetime.year
print("Current Year:", current_year)

month_of_year = current_datetime.strftime("%B")
print("Month of Year:", month_of_year)

week_number = current_datetime.strftime("%U")
print("Week Number of the Year:", week_number)

weekday_of_week = current_datetime.weekday()
print("Weekday of the Week:", weekday_of_week)


day_of_year = current_datetime.strftime("%j")
print("Day of Year:", day_of_year)


day_of_month = current_datetime.day
print("Day of Month:", day_of_month)


day_of_week = current_datetime.strftime("%w")
print("Day of Week:", day_of_week)
```

### 80. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries
Hint: First we need to find top-3 countries based on revenue, then create scater plot between Quantity and Unitprice for each country separately


```python

country_revenue = online_rt.groupby('Country')[['Quantity', 'UnitPrice']].sum()
country_revenue['Revenue'] = country_revenue['Quantity'] * country_revenue['UnitPrice']

top_countries = country_revenue.nlargest(3, 'Revenue').index

for country in top_countries:
    country_data = online_rt[(online_rt['Country'] == country) & (online_rt['Quantity'] > 0)]
    plt.figure(figsize=(10, 6))
    plt.scatter(country_data['UnitPrice'], country_data['Quantity'], alpha=0.5)
    plt.title(f"Quantity per UnitPrice Scatterplot - {country}")
    plt.xlabel("UnitPrice")
    plt.ylabel("Quantity")
    plt.show()
```
